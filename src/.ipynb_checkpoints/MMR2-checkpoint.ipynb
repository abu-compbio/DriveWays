```python
import numpy as np
from glob import glob
import pandas as pd
from scipy import stats
from pprint import pprint
import networkx as nx
from networkx.algorithms import matching
from tqdm import tqdm_notebook as tqdm
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def txt_parser2(file):

    with open(file) as f:
        lines = f.readlines()
        modules = [l.strip().split() for l in lines]

    return modules

def predict_modules(file):
    with open(file) as f:
        lines = f.readlines()
    modules = [l.strip().split() for l in lines]
    modules_ = []
    mapping = {}
    index = 0
    for i,l in enumerate(modules):
        if len(l) >= 2000:
            # np.random.shuffle(l)
            # l = list(l)
            mapping[index] = '{}_1'.format(i)
            mapping[index+1] = '{}_2'.format(i)
            index +=2
            modules_.extend([l[:len(l)//2], l[len(l)//2:]])
        else:
            mapping[index] = '{}'.format(i)
            index +=1
            modules_.append(l)
    return modules_, mapping

def cosmic_genes():
    fhinput = open('../data/Census_allTue_May_23_12-08-15_2017.tsv')
    cosmic_genes = []
    line = fhinput.readline()
    for line in fhinput:
        cosmic_genes.append(line.split()[0])
    return cosmic_genes

def bipartite(ref_modules,pred_modules):
    i = 0
    best_match_sum = 0
    miss_match = 0
    nodes = {i:s for i,s in enumerate(ref_modules+pred_modules)}
    edges = []
    n,m = len(ref_modules), len(pred_modules)
    for i,P in enumerate(pred_modules):
        best_ = 0
        for j,R in enumerate(ref_modules):
            u,v = j, i+len(ref_modules)
            intes = len(set(R).intersection(set(P)))
            w = intes
            if w > best_:
                best_ = w
            if w > 0:
                edges.append((u,v,w))
        if best_ == 0:
            miss_match += 1
        best_match_sum += best_
        
    best_match_avg = best_match_sum/min(n,m) 
    return nodes, edges,best_match_sum,best_match_avg, miss_match


def plots(ds_, ts_, arr, name='Score'):
    arr = np.array(arr)
    plot_ds = True
    axis_font = {'fontname':'Times New Roman', 'size':'15'}
    
    if isinstance(ds_, tuple):
        s,e = ds_
    if isinstance(ts_, tuple):
        s,e = ts_
        plot_ds = False
    if isinstance(ds_, tuple) and isinstance(ts_, tuple) :
        raise 'boht ds_ and ts_ cannont be tuple'
    legend_ = []  
    if plot_ds:
        fig = plt.figure()
        ax = fig.add_subplot(111)
        for i,r in enumerate(arr[s:e, :]):
            legend_.append('d = {}'.format(ds[s+i]))
            ax.plot(ts, r)
        legend = plt.legend(legend_, loc=8,fancybox=True, fontsize= 'large', framealpha=0,
                    edgecolor = 'b', ncol= 4, bbox_to_anchor=(0.5,-0.5))
        frame = legend.get_frame()
        plt.xlabel('t_threshold', **axis_font)
        plt.ylabel(name, **axis_font)
        plt.grid()
        plt.show()
    else:
        fig = plt.figure()
        ax = fig.add_subplot(111)
        for i,r in enumerate(arr.T[s:e, :]):
            legend_.append('t = {}'.format(ts[s+i]))
            ax.plot(ds, r)
        legend = plt.legend(legend_, loc=8,fancybox=True, fontsize= 'large', framealpha=0,
                    edgecolor = 'b', ncol= 4, bbox_to_anchor=(0.5,-0.5))
        frame = legend.get_frame()
        plt.xlabel('d_threshold', **axis_font)
        plt.ylabel(name, **axis_font)
        plt.grid()
#         plt.show()
    return ax
```

## Kegg

```python
data_path = '../data/'

comp_folder = '../out/components/24_06_2019/'

ds = [ 0,  1, 2, 3, 4,  5]
ts = [0.32, 0.44, 0.56, 0.68, 0.8 , 0.92, 1.04, 1.16, 1.28, 1.4 , 1.52, 1.64, 1.76, 1.88, 2.  ]



ref_modules = txt_parser2(data_path+'kegg6.2_02_05_2019.txt')
print('There are {} reference modules'.format(len(ref_modules)))

ref_genes = []
for m in ref_modules:
    ref_genes.extend(m)
ref_genes = set(ref_genes)
    



mmrs = []
total_matching = []
best_mmr = []
n_modules=[]
avg_modules_s = []
n_unique_genes = []
n_inters_genes = []

for d in ds:
    mmr= []
    t_m = []
    best_m = []
    n_m = []
    avg_s = []
    n_unique = []
    n_inters = []
    
    for t in ts:

        comp_file = 'd{}_t{}/cc_n1771_k3.txt'.format(d,t )
        pred_modules, mapping = predict_modules(comp_folder+comp_file)
        #print('There are {} predicted modules'.format(len(pred_modules)))
        n_m.append(len(pred_modules))
        
        genes = []
        for m in pred_modules:
            genes.extend(m)
        n_unique.append(len(set(genes)))
        
        n_inters.append(len(set(genes).intersection(ref_genes)))
        
        avg = sum([len(s) for s in pred_modules])/len(pred_modules)
        avg_s.append(avg)
        
        nodes, edges,best_match, miss_match = bipartite(ref_modules,pred_modules)
        #print('best match ', best_match, 'miss match ', miss_match)
        best_m.append(best_match)

        n,m = len(ref_modules), len(pred_modules)
        G= nx.Graph()
        G.add_weighted_edges_from(edges)
        G = G.to_undirected()

        max_match = matching.max_weight_matching(G) #, maxcardinality=True
        sum_ = 0

        for e in max_match:

            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_ += G[e[0]][e[1]]['weight']

        # mmr = sum_/len(max_match)
        mmr.append(sum_/min(n,m))

        zero_count = 0
        tuples = []
        n = len(ref_modules)
        for e in max_match:

            e0 = e[0] if e[0] < n else e[1]
            e1 = e[1] if e[1] >= n else e[0]
            # try:
            R = ref_modules[e0]
            
            try:
                P = pred_modules[e1-n]
            except:
                print('\t ',e1-n, )
                break
            p_idx = mapping[e1-n]
            inters = len(set(R).intersection(set(P)))
            if inters == 0: zero_count += 1

        t_m.append(len(max_match))
        # print('total matching: {}, zero count: {}'.format(len(max_match), zero_count))
    total_matching.append(t_m)
    mmrs.append(mmr)
    best_mmr.append(best_m)
    n_modules.append(n_m)
    avg_modules_s.append(avg_s)
    n_unique_genes.append(n_unique)
    n_inters_genes.append(n_inters)
    print(d)
```

```python
data_path = '../data/'

comp_folder = '../out/components/04_11_2019/'


ref_modules = txt_parser2(data_path+'kegg6.2_02_05_2019.txt')
print('There are {} reference modules'.format(len(ref_modules)))

ref_genes = []
for m in ref_modules:
    ref_genes.extend(m)
ref_genes = set(ref_genes)

mmrs = []
best_mmrs = []
n_predicted_modules = []
for i in tqdm(range(100)):
    comp_file = f'{i}/cc_n1771_k3.txt'
    pred_modules, mapping = predict_modules(comp_folder+comp_file)
    #print('There are {} predicted modules'.format(len(pred_modules)))
    n_predicted_modules.append(len(pred_modules))
    
    nodes, edges,best_match, miss_match = bipartite(ref_modules,pred_modules)

    best_mmrs.append(best_match)

    n,m = len(ref_modules), len(pred_modules)
    G= nx.Graph()
    G.add_weighted_edges_from(edges)
    G = G.to_undirected()

    max_match = matching.max_weight_matching(G, maxcardinality=True)
    sum_ = 0

    for e in max_match:

        assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
        sum_ += G[e[0]][e[1]]['weight']

    # mmr = sum_/len(max_match)
    mmr = sum_/min(n,m)

    
    mmrs.append(mmr)

print('Avg Best Matching: {}, Avg MMR score: {} '.format(np.mean(best_mmrs), np.mean(mmrs) ))
print(f'Avg # of Predicted Modules {np.mean(n_predicted_modules)}')
```

```python
np.save('../out/components/04_11_2019/avg_mmrs_intAct_Kegg', np.array(mmrs) )
```

```python
data_path = '../data/'

comp_folder = '../out/components/04_11_2019/'


ref_modules = txt_parser2(data_path+'kegg6.2_02_05_2019.txt')
print('There are {} reference modules'.format(len(ref_modules)))

ref_genes = []
for m in ref_modules:
    ref_genes.extend(m)
ref_genes = set(ref_genes)

ODMSS_scores = []
NCG_overlap_ratios = []
avg_module_sizes = []
for i in tqdm(range(100)):
    comp_file = f'{i}/cc_n1771_k3.txt'

    pred_modules, mapping = predict_modules(comp_folder+comp_file)
#         n_m.append(len(pred_modules))

    pred_genes = []
    for m in pred_modules:
        pred_genes.extend(m)
    pred_genes = set(pred_genes)


    sum_ = 0
    avg_size = []
    for pred_m in pred_modules:
        sum_ += mexcov.getcov(pred_m)* mexcov.getmutex(pred_m)
        avg_size.append(len(pred_m))
#     avg_ = sum_ /len(pred_modules)

    if np.mean(avg_size) > 50:
        avg_module_sizes.append(50)
    else:
        avg_module_sizes.append(np.mean(avg_size))

    ODMSS_scores.append(sum_)

    NCG_overlap_ratios.append(len(set(ref_genes).intersection(pred_genes))/len(pred_genes))
```

```python
from tqdm import tqdm_notebook as tqdm
data_path = '../data'
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n1771_k3','cc_n3368_k3','cc_n1173_k3']
ref_names = ['kegg6.2_02_05_2019','Reactom5.0_02_05_2019','Biocarta5.0_02_05_2019']
comp_folder = '../out/components/04_11_2019/'

with open('../data/intact_index_file_theresholded0_35.txt') as f:
    intAct_genes = [s.strip().split('\t')[-1] for s in f.readlines()]
    



for path,ref_name,pred_name, in zip(pathways,ref_names,pred_names):
    
    mmr_qvals = []
    mmr_EFs = []
    ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt')
    
    for i in tqdm(range(100)):
        
        pred_modules, mapping = predict_modules(f'{comp_folder}/{i}/{pred_name}.txt')
        nodes, edges,best_match_sum,best_match_avg, miss_match = bipartite(ref_modules,pred_modules)
        n_ref, n_pred =len(ref_modules), len(pred_modules)


        M = len(intAct_genes)
        pvals = []
        for i in range(len(ref_modules)):
            for j in range(len(pred_modules)):
                a = ref_modules[i]
                b = pred_modules[j]
                n = len(a)
                N = len(b)
                x = len(set(a).intersection(set(b)))
                pvals.append(hypergeom.sf(x-1, M, n, N))

        qvals = multitest.multipletests(pvals, alpha=0.05, method='fdr_bh', is_sorted=False, returnsorted=False)

        edges_pvals = []
        edges_EF = []        
        reject = qvals[0]
        corr_pvals = qvals[1]
        enriched_tuples = {}
        for i in range(len(ref_modules)):
            for j in range(len(pred_modules)):
                if reject[i*len(pred_modules) + j] == True:
                    w = -np.log10(corr_pvals[i*len(pred_modules) + j])
                    edges_pvals.append((j, i+n_pred+1,w))

                    a = ref_modules[i]
                    b = pred_modules[j]
                    n = len(a)
                    N = len(b)
                    x = len(set(a).intersection(set(b)))
                    s = len(set(a).intersection(set(intAct_genes)))
                    enrich_factor = (x/s)*(M/N)
                    edges_EF.append((j, i+n_pred+1,enrich_factor))


        G= nx.Graph()
        G.add_weighted_edges_from(edges_pvals)
        G = G.to_undirected()
        max_match = matching.max_weight_matching(G) #, maxcardinality=True

        sum_pvals= 0
        for e in max_match:
            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_pvals += G[e[0]][e[1]]['weight']

        G= nx.Graph()
        G.add_weighted_edges_from(edges_EF)
        G = G.to_undirected()
        max_match = matching.max_weight_matching(G) #, maxcardinality=True

        sum_EF= 0
        for e in max_match:
            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_EF += G[e[0]][e[1]]['weight']

        mmr_qvals.append(sum_pvals/min(n_ref, n_pred))
        mmr_EFs.append(sum_EF/min(n_ref, n_pred))


    np.save(f'../out/components/04_11_2019/mmr_qvals_{path}', np.array(mmr_qvals) )
    np.save(f'../out/components/04_11_2019/mmr_EFs_{path}', np.array(mmr_EFs) )
```
