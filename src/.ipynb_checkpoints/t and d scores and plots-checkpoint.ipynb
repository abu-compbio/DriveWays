```python
def txt_parser2(file):

    with open(file) as f:
        lines = f.readlines()
        modules = [l.strip().split() for l in lines]

    return modules

def predict_modules(file):
    with open(file) as f:
        lines = f.readlines()
    modules = [l.strip().split() for l in lines]
    modules_ = []
    mapping = {}
    index = 0
    for i,l in enumerate(modules):
        if len(l) >= 2000:
            # np.random.shuffle(l)
            # l = list(l)
            mapping[index] = '{}_1'.format(i)
            mapping[index+1] = '{}_2'.format(i)
            index +=2
            modules_.extend([l[:len(l)//2], l[len(l)//2:]])
        else:
            mapping[index] = '{}'.format(i)
            index +=1
            modules_.append(l)
    return modules_, mapping

def cosmic_genes():
    fhinput = open('../data/Census_allTue_May_23_12-08-15_2017.tsv')
    cosmic_genes = []
    line = fhinput.readline()
    for line in fhinput:
        cosmic_genes.append(line.split()[0])
    return cosmic_genes

def bipartite(ref_modules,pred_modules):
    i = 0
    best_match_sum = 0
    miss_match = 0
    nodes = {i:s for i,s in enumerate(ref_modules+pred_modules)}
    edges = []
    n,m = len(ref_modules), len(pred_modules)
    for i,P in enumerate(pred_modules):
        best_ = 0
        for j,R in enumerate(ref_modules):
            u,v = j, i+len(ref_modules)
            intes = len(set(R).intersection(set(P)))
            w = intes
            if w > best_:
                best_ = w
            if w > 0:
                edges.append((u,v,w))
        if best_ == 0:
            miss_match += 1
        best_match_sum += best_
        
    best_match_avg = best_match_sum/min(n,m) 
    return nodes, edges,best_match_sum,best_match_avg, miss_match






import matplotlib as mpl

def plots(ds_, ts_, arr, ax, ylabel='Score', xlabel = 'param?', title='TITLE' ):
    colors = colourWheel[:5]+colourWheel[9:]
    arr = np.array(arr)
    plot_ds = True
    
    if isinstance(ds_, tuple):
        s,e = ds_
    if isinstance(ts_, tuple):
        s,e = ts_
        plot_ds = False
    if isinstance(ds_, tuple) and isinstance(ts_, tuple) :
        raise 'boht ds_ and ts_ cannont be tuple'
    legend_ = []  
    if plot_ds:
#         fig = plt.figure()
#         ax = fig.add_subplot(111)
        for i,r in enumerate(arr[s:e, :]):
            sns.lineplot(ts, r, ax = ax,legend= 'brief', label ='d = {}'.format(ds[s+i]), color = colors[i])

#         plt.legend(loc=0, ncol=2)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_title(title)    
        ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(5))
        ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(5))
        ax.grid(True)
    else:
#         fig = plt.figure()
#         ax = fig.add_subplot(111)
        for i,r in enumerate(arr.T[s:e, :]):
            sns.lineplot(ds, r, ax = ax, legend= 'brief', label ='t = {}'.format(ts[s+i]), color = colors[i])

#         plt.legend(loc=0, ncol=2)
        ax.set_xlabel(xlabel)
        ax.set_ylabel(ylabel)
        ax.set_title(title)    
        ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(5))
        ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(5))
        ax.grid(True)
    ax.get_legend().set_visible(False)
    return ax
```

```python
data_path = '../data/'

comp_folder = '../out/components/24_06_2019/'

ds = [ 0,  1, 2, 3, 4,  5]
ts = [0.32, 0.44, 0.56, 0.68, 0.8 , 0.92, 1.04, 1.16, 1.28, 1.4 , 1.52, 1.64, 1.76, 1.88, 2.  ]



ref_modules = txt_parser2(data_path+'kegg6.2_02_05_2019.txt')
print('There are {} reference modules'.format(len(ref_modules)))

ref_genes = []
for m in ref_modules:
    ref_genes.extend(m)
ref_genes = set(ref_genes)
    




ODMSS_scores = []
NCG_overlap_ratios = []
avg_module_sizes = []
for d in ds:
    ODMSS_score = []
    NCG_overlap_ratio = []
    avg_module_size = []
    
    for t in ts:

        comp_file = 'd{}_t{}/cc_n1771_k3.txt'.format(d,t )
        pred_modules, mapping = predict_modules(comp_folder+comp_file)
#         n_m.append(len(pred_modules))
        
        pred_genes = []
        for m in pred_modules:
            pred_genes.extend(m)
        pred_genes = set(pred_genes)
        
        
        sum_ = 0
        avg_size = []
        for pred_m in pred_modules:
            sum_ += mexcov.getcov(pred_m)* mexcov.getmutex(pred_m)
            avg_size.append(len(pred_m))
#         avg_ = sum_ /len(pred_modules)
    
        if np.mean(avg_size) > 50:
            avg_module_size.append(50)
        else:
            avg_module_size.append(np.mean(avg_size))
            
        ODMSS_score.append(sum_)
        
        NCG_overlap_ratio.append(len(set(ref_genes).intersection(pred_genes))/len(pred_genes))
        
    ODMSS_scores.append(ODMSS_score)
    NCG_overlap_ratios.append(NCG_overlap_ratio)
    avg_module_sizes.append(avg_module_size) 
    print(d)   
        
```
