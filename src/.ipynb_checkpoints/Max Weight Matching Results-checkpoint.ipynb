```python
def txt_parser2(file):

    with open(file) as f:
        lines = f.readlines()
        modules = [l.strip().split() for l in lines]

    return modules

def predict_modules(file):
    with open(file) as f:
        lines = f.readlines()
    modules = [l.strip().split() for l in lines]
    modules_ = []
    mapping = {}
    index = 0
    for i,l in enumerate(modules):
        if len(l) >= 2000:
            # np.random.shuffle(l)
            # l = list(l)
            mapping[index] = '{}_1'.format(i)
            mapping[index+1] = '{}_2'.format(i)
            index +=2
            modules_.extend([l[:len(l)//2], l[len(l)//2:]])
        else:
            mapping[index] = '{}'.format(i)
            index +=1
            modules_.append(l)
    return modules_, mapping


def bipartite(ref_modules,pred_modules):
    i = 0
    best_match_sum = 0
    miss_match = 0
    nodes = {i:s for i,s in enumerate(ref_modules+pred_modules)}
    edges = []
    n,m = len(ref_modules), len(pred_modules)
    for i,P in enumerate(pred_modules):
        best_ = 0
        for j,R in enumerate(ref_modules):
            u,v = j, i+len(ref_modules)
            intes = len(set(R).intersection(set(P)))
            w = intes
            if w > best_:
                best_ = w
            if w > 0:
                edges.append((u,v,w))
        if best_ == 0:
            miss_match += 1
        best_match_sum += best_
        
    best_match_avg = best_match_sum/min(n,m) 
    return nodes, edges,best_match_sum,best_match_avg, miss_match


```

```python
data_path = '../data'
models = ['ClusterOne', 'mexcogrowth', 'hotnet2', 'memcover', 'mexcowalk']
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n1771_k3','cc_n3368_k3','cc_n1173_k3']

comp_folder = '../out/components/final_results_10_10_2019'
ref_names = ['kegg6.2_02_05_2019','Reactom5.0_02_05_2019','Biocarta5.0_02_05_2019']


methods = []
reference =[]
mmr_qval = []
mmr_EF = []

for model in models:
    for path,ref_name,pred_name, in zip(pathways,ref_names,pred_names):
        ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt')
        pred_modules, mapping = predict_modules(f'{comp_folder}/{model}/{pred_name}.txt')
        nodes, edges,best_match_sum,best_match_avg, miss_match = bipartite(ref_modules,pred_modules)
        n_ref, n_pred =len(ref_modules), len(pred_modules)
        
        
        M = len(background_genes)
        pvals = []
        for i in range(len(ref_modules)):
            for j in range(len(pred_modules)):
                a = ref_modules[i]
                b = pred_modules[j]
                n = len(a)
                N = len(b)
                x = len(set(a).intersection(set(b)))
                pvals.append(hypergeom.sf(x-1, M, n, N))
        
        qvals = multitest.multipletests(pvals, alpha=0.05, method='fdr_bh', is_sorted=False, returnsorted=False)

        edges_pvals = []
        edges_EF = []        
        reject = qvals[0]
        corr_pvals = qvals[1]
        enriched_tuples = {}
        for i in range(len(ref_modules)):
            for j in range(len(pred_modules)):
                if reject[i*len(pred_modules) + j] == True:
                    w = -np.log10(corr_pvals[i*len(pred_modules) + j])
                    edges_pvals.append((j, i+n_pred+1,w))
                    
                    a = ref_modules[i]
                    b = pred_modules[j]
                    n = len(a)
                    N = len(b)
                    x = len(set(a).intersection(set(b)))
                    s = len(set(a).intersection(set(intAct_genes)))
                    enrich_factor = (x/s)*(M/N)
                    edges_EF.append((j, i+n_pred+1,enrich_factor))
        
        
        G= nx.Graph()
        G.add_weighted_edges_from(edges_pvals)
        G = G.to_undirected()
        max_match = matching.max_weight_matching(G) #, maxcardinality=True
        
        sum_pvals= 0
        for e in max_match:
            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_pvals += G[e[0]][e[1]]['weight']
        
        G= nx.Graph()
        G.add_weighted_edges_from(edges_EF)
        G = G.to_undirected()
        max_match = matching.max_weight_matching(G) #, maxcardinality=True
        
        sum_EF= 0
        for e in max_match:
            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_EF += G[e[0]][e[1]]['weight']
        
        mmr_qval.append(sum_pvals/min(n_ref, n_pred))
        mmr_EF.append(sum_EF/min(n_ref, n_pred))
        methods.append(model)
        reference.append(path)
```

```python
d = { 'MMR qval':mmr_qval ,
     'MMR EF':mmr_EF ,
     'Method':methods,
     'Reference' : reference,
}
df_mmr_enrichment = pd.DataFrame(data=d)
df_mmr_enrichment['Method'] = df_mmr_enrichment['Method'].apply(lambda x: r'$\mathbf{OLDRIM}$' if x == 'mexcogrowth' else x)
df_mmr_enrichment['Method'] = df_mmr_enrichment['Method'].apply(lambda x: 'Hotnet2' if x == 'hotnet2' else x)
df_mmr_enrichment['Method'] = df_mmr_enrichment['Method'].apply(lambda x: 'MEMCover' if x == 'memcover' else x)
df_mmr_enrichment['Method'] = df_mmr_enrichment['Method'].apply(lambda x: 'MEXCOWalk' if x == 'mexcowalk' else x)

df_mmr_enrichment['Reference'] = df_mmr_enrichment['Reference'].apply(lambda x: 'KEGG' if x == 'kegg' else x)
df_mmr_enrichment['Reference'] = df_mmr_enrichment['Reference'].apply(lambda x: 'Reactome' if x == 'reactom' else x)
df_mmr_enrichment['Reference'] = df_mmr_enrichment['Reference'].apply(lambda x: 'BioCarta' if x == 'biocarta' else x)
```

```python
data_path = '../data'
models = ['ClusterOne', 'mexcogrowth', 'hotnet2', 'memcover', 'mexcowalk']
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n1771_k3','cc_n3368_k3','cc_n1173_k3']

comp_folder = '../out/components/final_results_10_10_2019'
ref_names = ['kegg6.2_02_05_2019','Reactom5.0_02_05_2019','Biocarta5.0_02_05_2019']


methods = []
reference =[]
best_scores_sum = []
best_scores_avg = []
mwm_scores_sum = []
mwm_score_avg = []
for model in models:
    for path,ref_name,pred_name, in zip(pathways,ref_names,pred_names):
        ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt')
        pred_modules, mapping = predict_modules(f'{comp_folder}/{model}/{pred_name}.txt')
        nodes, edges,best_match_sum,best_match_avg, miss_match = bipartite(ref_modules,pred_modules)
        n_ref, m =len(ref_modules), len(pred_modules)
        file = f'../out/Expander-pathway Enrichment/{model}/{path}_enrichment_fdr_allgroups.txt'
        res_df = pd.read_csv(file, sep='\t')
        edges = []
        edges1 = []
        best_edges = {}
        pred_modules = set()
        ref_modules = set()
        n = max(res_df['Set'].values)+1
        for r in res_df.iterrows():
#             print(r)
#             print(r[1], len(r[1]), type(r[1]))
            e0 = str(r[1]['Set'])
            e1 = str(int(r[1]['Enriched with'].split(' ')[-1])+n)
######## Pay attention to w and w1. the param which is assigned to w is the one that we are getting the results for         
            w1 = int(r[1]['Enrichment factor'])
            w = -np.log10(float(r[1]['Corrected p-value']))
#             print(f'{e0}, {e1}, {w}')
            edges.append((e0,e1,w))
            edges1.append((e0,e1,w1))

            if e0 in best_edges.keys():
                if w > best_edges[e0]:
                    best_edges[e0] = w
            else:
                best_edges[e0] = w
        best_score_sum = sum(best_edges.values())
        best_score_avg = best_score_sum/ len(best_edges.values())
        
        
        G= nx.Graph()
        G.add_weighted_edges_from(edges)
        G = G.to_undirected()
        max_match = matching.max_weight_matching(G) #, maxcardinality=True
        
        G1= nx.Graph()
        G1.add_weighted_edges_from(edges1)
        G1 = G1.to_undirected()
        max_match1 = matching.max_weight_matching(G1) #, maxcardinality=True
        inters = set(max_match1).intersection(max_match)

        sum_= 0
        for e in max_match:
            assert G[e[0]][e[1]]['weight'] == G[e[1]][e[0]]['weight'], 'something is wrong'
            sum_ += G[e[0]][e[1]]['weight']
#         print(f'model = {model}, pathway = {path}')
#         print(f'Best Score = {best_score:.4f}, MWM = {sum_:.4f}, Avg_MWM = {sum_/len(max_match):.4f}, #Metching Edges = {len(max_match)}\n')
        methods.append(model)
        reference.append(path)
        best_scores_sum.append(best_score_sum)
        best_scores_avg.append(best_score_avg)
        mwm_scores_sum.append(sum_)
        mwm_score_avg.append(sum_/min(n_ref,m ))
```
