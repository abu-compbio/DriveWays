```python
def predict_modules(file):
    with open(file) as f:
        lines = f.readlines()
    modules = [l.strip().split() for l in lines]
    modules_ = []
    mapping = {}
    index = 0
    for i,l in enumerate(modules):
        if len(l) >= 2000:
            # np.random.shuffle(l)
            # l = list(l)
            mapping[index] = '{}_1'.format(i)
            mapping[index+1] = '{}_2'.format(i)
            index +=2
            modules_.extend([l[:len(l)//2], l[len(l)//2:]])
        else:
            mapping[index] = '{}'.format(i)
            index +=1
            modules_.append(l)
    return modules_, mapping
def txt_parser2(file):

    with open(file) as f:
        lines = f.readlines()
        modules = [l.strip().split() for l in lines]

    return modules
```

```python
from CovMex import *
mexcov = CovMex('../data/gene_patient_BRCA.txt')
```

```python
data_path = '../data'
models = ['ClusterOne', 'mexcogrowth', 'hotnet2', 'memcover', 'mexcowalk']
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n720_k3','cc_n957_k3','cc_n418_k3']

comp_folder = '../out/components/BRCA'
ref_names = ['kegg6.2_BRCA','Reactom5.0_BRCA','Biocarta5.0_BRCA']

methods = []
reference =[]
ODMSS_scores = []
NCG_overlap_ratios = []

for model in models:
    
    for path,ref_name,pred_name, in zip(pathways,ref_names,pred_names):
        
        
        ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt')

        ref_genes = []
        for m in ref_modules:
            ref_genes.extend(m)
        ref_genes = set(ref_genes)
        

        pred_modules, mapping = predict_modules(f'{comp_folder}/{model}/{pred_name}.txt')
        pred_genes = []
        for m in pred_modules:
            pred_genes.extend(m)
        pred_genes = set(pred_genes)

        avg_module_sizes = []
        sum_ = 0
        avg_size = []
        for pred_m in pred_modules:
            sum_ += mexcov.getcov(pred_m)* mexcov.getmutex(pred_m)
            avg_size.append(len(pred_m))
        
#         avg_ = sum_ /len(pred_modules)
        
        methods.append(model)
        reference.append(path)
#         avg_module_size.append(np.mean(avg_size))
        ODMSS_scores.append(sum_)
        NCG_overlap_ratios.append(len(set(ref_genes).intersection(pred_genes))/len(pred_genes))
```

```python
d = { 'ODMSS': ODMSS_scores,
     'NCG overlap': NCG_overlap_ratios,
     'Method': methods,
     'Reference' : reference,

}
df = pd.DataFrame(data=d)

df['Method'] = df['Method'].apply(lambda x: r'$\mathbf{OLDRIM}$' if x == 'mexcogrowth' else x)
df['Method'] = df['Method'].apply(lambda x: 'Hotnet2' if x == 'hotnet2' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEMCover' if x == 'memcover' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEXCOWalk' if x == 'mexcowalk' else x)

df['Reference'] = df['Reference'].apply(lambda x: 'KEGG' if x == 'kegg' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'Reactome' if x == 'reactom' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'BioCarta' if x == 'biocarta' else x)


# df.head()
```

```python
f, axes = plt.subplots(1, 2, figsize= (7.2*2,4.45))

ax = sns.barplot(x="Reference", y="ODMSS", hue="Method", data=df, ax=axes[1])
ax.set_xlabel('')
ax.set_ylabel("ODMSS")
# ax.set_title('DMS Scores')
ax.get_legend().set_visible(False)
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(4))
ax.grid(False)
ax.yaxis.grid(True)

ax.set_xticklabels(['720\nKEGG', '957\nReactome', '418\nBioCarta'])

ax = sns.barplot(x="Reference", y="NCG overlap", hue="Method", data=df, ax=axes[0])
ax.set_xlabel("")
ax.set_ylabel("NCG Breast Cancer Overlap Ratio", fontsize=14)
ax.get_legend().set_visible(False)
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator(4))
ax.grid(False)
ax.yaxis.grid(True)
sns.despine()


lines_labels = [ax.get_legend_handles_labels()]
lines, labels = [sum(lol, []) for lol in zip(*lines_labels)]

legend = f.legend( lines, labels, ncol = 1,loc=7, bbox_transform = plt.gcf().transFigure, 
          bbox_to_anchor=(1.03, 0.78,)
          )
art = [legend]

for n, ax in enumerate(axes):
    ax.text(-0.1, 1.05, string.ascii_uppercase[n], transform=ax.transAxes, 
            size=20, weight='bold')
sns.despine()
# plt.legend()
```

```python
data_path = '../data'
pathways =  ['KEGG', 'Reactome', 'BioCarta']
ref_names = ['kegg6.2_BRCA','Reactom5.0_BRCA','Biocarta5.0_BRCA']

reference =[]
pathway_avg_module_size = []
for ref_name in ref_names:
    ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt') 
    avg_size = []
    for ref_m in ref_modules:
        avg_size.append(len(ref_m))
    pathway_avg_module_size.append(np.mean(avg_size))
    
```

```python
data_path = '../data'
models = ['ClusterOne', 'mexcogrowth', 'hotnet2', 'memcover', 'mexcowalk', 'Reference']
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n720_k3','cc_n957_k3','cc_n418_k3']

comp_folder = '../out/components/BRCA'
ref_names = ['kegg6.2_BRCA','Reactom5.0_BRCA','Biocarta5.0_BRCA']

# this is set to get the results for mexcogrowth that outputs # of modules similar to ref pathways
# models = [ 'mexcogrowth']
# pred_names = ['cc_n1771_k3 98_modules','cc_n3368_k3 325_modules','cc_n1173_k3 139_modules']


methods = []
reference =[]

avg_module_size = []
for model in models:
    if model == 'Reference':
        avg_module_size.extend(pathway_avg_module_size)
        methods.extend(['Reference']*3)
        reference.extend(pathways)
        break
        
    for path,ref_name,pred_name, in zip(pathways,ref_names,pred_names):
        pred_modules, mapping = predict_modules(f'{comp_folder}/{model}/{pred_name}.txt')

        avg_size = []
        for pred_m in pred_modules:
            avg_size.append(len(pred_m))

        
        methods.append(model)
        reference.append(path)
#         if np.mean(avg_size) > 50:
#             avg_module_size.append(52)
#         else:
        avg_module_size.append(np.mean(avg_size))

d = {'Method': methods,
     'Reference' : reference,
     'avg Module size': avg_module_size
}
df = pd.DataFrame(data=d)

df['Method'] = df['Method'].apply(lambda x: r'$\mathbf{OLDRIM}$' if x == 'mexcogrowth' else x)
df['Method'] = df['Method'].apply(lambda x: 'Hotnet2' if x == 'hotnet2' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEMCover' if x == 'memcover' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEXCOWalk' if x == 'mexcowalk' else x)

df['Reference'] = df['Reference'].apply(lambda x: 'KEGG' if x == 'kegg' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'Reactome' if x == 'reactom' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'BioCarta' if x == 'biocarta' else x)    
```

```python
data_path = '../data'
pathways =  ['KEGG', 'Reactome', 'BioCarta']
ref_names = ['kegg6.2_BRCA','Reactom5.0_BRCA','Biocarta5.0_BRCA']

reference =[]
avg_module_size = []
pathwyas_overlap = []
for ref_name,pathway in zip(ref_names,pathways) :
    ref_modules = txt_parser2(f'{data_path}/{ref_name}.txt') 
    overlaps = []
    for i in range(len(ref_modules[:-1])):
        for j in range(i, len(ref_modules[:-1])):
            m,n= ref_modules[i], ref_modules[j]
            w= len(set(n).intersection(set(m)))**2/(len(n)*len(m))
            overlaps.append(w)
    pathwyas_overlap.append(np.mean(overlaps))
    print(f'The pairwise overlap for {pathway} is {np.mean(overlaps)}')
```

```python
data_path = '../data'
models = ['ClusterOne', 'mexcogrowth', 'memcover', 'Reference']
pathways =  ['kegg', 'reactom', 'biocarta']
pred_names = ['cc_n720_k3','cc_n957_k3','cc_n418_k3']

comp_folder = '../out/components/BRCA'
ref_names = ['kegg6.2_BRCA','Reactom5.0_BRCA','Biocarta5.0_BRCA']
# this is set to get the results for mexcogrowth that outputs # of modules similar to ref pathways
# models = [ 'mexcogrowth']
# pred_names = ['cc_n1771_k3 98_modules','cc_n3368_k3 325_modules','cc_n1173_k3 139_modules']


methods = []
reference =[]
pairwise_overlap = []
for model in models:
    if model == 'Reference':
        pairwise_overlap.extend(pathwyas_overlap)
        methods.extend(['Reference']*3)
        reference.extend(pathways)
        break
        
    for path,ref_name,pred_name in zip(pathways,ref_names,pred_names):
        pred_modules, mapping = predict_modules(f'{comp_folder}/{model}/{pred_name}.txt')
        overlaps = []
        
        for i in range(len(pred_modules)):
            for j in range(i+1, len(pred_modules)):
                m,n= pred_modules[i], pred_modules[j]
                w= len(set(n).intersection(set(m)))**2/(len(n)*len(m))
                overlaps.append(w)
        methods.append(model)
        reference.append(path)
        pairwise_overlap.append(np.mean(overlaps))
       
```

```python
d = { 
     'Method': methods,
     'Reference' : reference,
     'pairwise_overlap': pairwise_overlap
}
df = pd.DataFrame(data=d)

df['Method'] = df['Method'].apply(lambda x: r'$\mathbf{OLDRIM}$' if x == 'mexcogrowth' else x)
df['Method'] = df['Method'].apply(lambda x: 'Hotnet2' if x == 'hotnet2' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEMCover' if x == 'memcover' else x)
df['Method'] = df['Method'].apply(lambda x: 'MEXCOWalk' if x == 'mexcowalk' else x)

df['Reference'] = df['Reference'].apply(lambda x: 'KEGG' if x == 'kegg' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'Reactome' if x == 'reactom' else x)
df['Reference'] = df['Reference'].apply(lambda x: 'BioCarta' if x == 'biocarta' else x)
df_pairwise_overlap = df
```
